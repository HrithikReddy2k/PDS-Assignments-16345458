```python
df = df.drop('New_Price', axis=1)
```

```python
df.head()
```

```python
def clean_and_extract_numeric(x):
    return x.str.extract('(\d+\.?\d*)', expand=False)

# Apply the function to the specified columns
df['Mileage'] = clean_and_extract_numeric(df['Mileage'])
df['Engine'] = clean_and_extract_numeric(df['Engine'])
df['Power'] = clean_and_extract_numeric(df['Power'])
```

```python
df.head()
```

```python
print(df.dtypes)
```

```python
#Converting object type to float type
df['Mileage'] = df['Mileage'].astype(float)
df['Engine'] = df['Engine'].astype(float)
df['Power'] = df['Power'].astype(float)
df['Seats']= df['Seats'].astype(float)
```

```python
df.head()
```

```python
mean_value_Mileage = df['Mileage'].mean()
mean_value_Engine = df['Engine'].mean()
mean_value_Seats = df['Seats'].mean()
mean_value_Power = df['Power'].mean()

# Replace null values with the mean
df['Mileage'].fillna(mean_value_Mileage, inplace=True)
df['Engine'].fillna(mean_value_Engine, inplace=True)
df['Seats'].fillna(mean_value_Seats, inplace=True)
df['Power'].fillna(mean_value_Power, inplace=True)
```

```python
df['Mileage'] = df['Mileage'].round(2)
```

```python
df['Engine'] = df['Engine'].astype(int)
```

```python
df['Seats'] = df['Seats'].astype(int)
```

```python
df['Power'] = df['Power'].round(2)
```

```python
# Perform one-hot encoding
df1 = pd.get_dummies(df, columns=['Fuel_Type', 'Transmission'])

# Display the DataFrame with one-hot encoded columns
print(df1)
```
