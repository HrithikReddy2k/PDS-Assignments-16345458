```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('diabetes.csv')
np.random.seed(42)
sample = data.sample(n=25)

population_mean_glucose = data['Glucose'].mean()
population_max_glucose = data['Glucose'].max()
sample_mean_glucose = sample['Glucose'].mean()
sample_max_glucose = sample['Glucose'].max()

print(f"Population Mean Glucose: {population_mean_glucose}")
print(f"Population Max Glucose: {population_max_glucose}")
print(f"Sample Mean Glucose: {sample_mean_glucose}")
print(f"Sample Max Glucose: {sample_max_glucose}")

labels = ['Population Mean', 'Population Max', 'Sample Mean', 'Sample Max']
values = [population_mean_glucose, population_max_glucose, sample_mean_glucose, sample_max_glucose]

plt.bar(labels, values, color=['blue', 'green', 'orange', 'red'])
plt.title('Comparison of Glucose Statistics (Population vs. Sample)')
plt.ylabel('Glucose Value')
plt.show()
```

```python
population_98th_percentile_bmi = np.percentile(data['BMI'], 98)
sample_98th_percentile_bmi = np.percentile(sample['BMI'], 98)

print(f"Population 98th Percentile BMI: {population_98th_percentile_bmi}")
print(f"Sample 98th Percentile BMI: {sample_98th_percentile_bmi}")

labels = ['Population 98th Percentile BMI', 'Sample 98th Percentile BMI']
values = [population_98th_percentile_bmi, sample_98th_percentile_bmi]

plt.bar(labels, values, color=['blue', 'orange'])
plt.title('Comparison of 98th Percentile BMI (Population vs. Sample)')
plt.ylabel('BMI Value')
plt.show()
```

```python
num_bootstrap_samples = 500
bootstrap_sample_size = 150

bootstrap_means = np.zeros(num_bootstrap_samples)
bootstrap_std_devs = np.zeros(num_bootstrap_samples)
bootstrap_percentiles = np.zeros(num_bootstrap_samples)

for i in range(num_bootstrap_samples):
    bootstrap_sample = np.random.choice(data['BloodPressure'], size=bootstrap_sample_size, replace=True)
    bootstrap_means[i] = np.mean(bootstrap_sample)
    bootstrap_std_devs[i] = np.std(bootstrap_sample)
    bootstrap_percentiles[i] = np.percentile(bootstrap_sample, 50)

population_mean_bp = data['BloodPressure'].mean()
population_std_dev_bp = data['BloodPressure'].std()
population_percentile_bp = np.percentile(data['BloodPressure'], 50)

print(f"Population Mean BloodPressure: {population_mean_bp}")
print(f"Population Standard Deviation BloodPressure: {population_std_dev_bp}")
print(f"Population 50th Percentile BloodPressure: {population_percentile_bp}")

average_bootstrap_mean = np.mean(bootstrap_means)
average_bootstrap_std_dev = np.mean(bootstrap_std_devs)
average_bootstrap_percentile = np.mean(bootstrap_percentiles)

print(f"Average Bootstrap Mean BloodPressure: {average_bootstrap_mean}")
print(f"Average Bootstrap Standard Deviation BloodPressure: {average_bootstrap_std_dev}")
print(f"Average Bootstrap 50th Percentile BloodPressure: {average_bootstrap_percentile}")

labels = ['Population Mean', 'Average Bootstrap Mean']
values = [population_mean_bp, average_bootstrap_mean]

plt.bar(labels, values, color=['blue', 'orange'])
plt.title('Comparison of Mean BloodPressure (Population vs. Bootstrap Samples)')
plt.ylabel('BloodPressure Value')
plt.show()
```

```python
labels = ['Population Std Dev', 'Average Bootstrap Std Dev']
values = [population_std_dev_bp, average_bootstrap_std_dev]

plt.bar(labels, values, color=['green', 'red'])
plt.title('Comparison of Std Dev BloodPressure (Population vs. Bootstrap Samples)')
plt.ylabel('BloodPressure Value')
plt.show()
```

```python
labels = ['Population 50th Percentile', 'Average Bootstrap 50th Percentile']
values = [population_percentile_bp, average_bootstrap_percentile]

plt.bar(labels, values, color=['purple', 'brown'])
plt.title('Comparison of 50th Percentile BloodPressure (Population vs. Bootstrap Samples)')
plt.ylabel('BloodPressure Value')
plt.show()
```

There is no much difference in between these two values
