```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_link = "/content/raw_yield_data.csv"
df = pd.read_csv(file_link, encoding = "unicode_escape")
```

```python
#renaming the data frame column names
df=df.set_axis(['Height', 'Weight', 'Age','Grip','Frailty'], axis='columns')
print(df)
print(df.dtypes)
```

```python
#convert Frailty to binary

from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()

df['Frailty']=label.fit_transform(df['Frailty'])
```

```python
import numpy as np
from google.colab import autoviz

def categorical_histogram(df, colname, figscale=1, mpl_palette_name='Dark2'):
  from matplotlib import pyplot as plt
  import seaborn as sns
  df.groupby(colname).size().plot(kind='barh', color=sns.palettes.mpl_palette(mpl_palette_name), figsize=(8*figscale, 4.8*figscale))
  plt.gca().spines[['top', 'right',]].set_visible(False)
  return autoviz.MplChart.from_current_mpl_state()

chart = categorical_histogram(df, *['Weight'], **{})
chart
```
